```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/telecom_traffic_data.csv")

hourly = df.groupby("Season")[["Arrivals", "Departures"]].mean()

hourly.plot(figsize=(12,6))
plt.title("Average Traffic by Season")
plt.xlabel("Season (1â€“4)")
plt.ylabel("Traffic")
plt.grid(True)
plt.show()
```
