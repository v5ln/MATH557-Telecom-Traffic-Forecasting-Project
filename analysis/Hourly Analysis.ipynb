```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/telecom_traffic_data.csv")

hourly = df.groupby("Period")[["Arrivals", "Departures"]].mean()

hourly.plot(figsize=(12,6))
plt.title("Average Traffic by Hour of Day")
plt.xlabel("Hour (1â€“24)")
plt.ylabel("Traffic")
plt.grid(True)
plt.show()
```
