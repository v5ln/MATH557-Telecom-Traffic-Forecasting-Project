```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/telecom_traffic_data.csv")

hourly = df.groupby("PricePerGB")[["Arrivals", "Departures"]].mean()

hourly.plot(figsize=(12,6))
plt.title("Average Traffic by Price (1 discounted, 2 regular price)")
plt.xlabel("Price (1â€“2)")
plt.ylabel("Traffic")
plt.grid(True)
plt.show()
```
