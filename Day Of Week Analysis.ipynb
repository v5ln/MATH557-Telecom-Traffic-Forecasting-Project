```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/telecom_traffic_data.csv")

hourly = df.groupby("DayOfWeek")[["Arrivals", "Departures"]].mean()

hourly.plot(figsize=(12,6))
plt.title("Average Traffic by Day of Week")
plt.xlabel("Day (1â€“7)")
plt.ylabel("Traffic")
plt.grid(True)
plt.show()
```
